```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
data_path = f"{path}/books.csv"
data_path
```

```python
import pandas as pd
```

```python
books = pd.read_csv(data_path)
books
```

```python
books.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax, cmap='GnBu')

plt.xlabel('Columns')
plt.ylabel('Missing Values')

plt.show()
```

```python
import numpy as np
```

```python
books['missing_description'] = np.where(books['description'].isna(), 1, 0)
books['age_of_book'] = 2025 - books['published_year']
```

```python
columns_of_interest = ['missing_description', 'age_of_book', 'average_rating', 'num_pages']

correlation_matrix = books[columns_of_interest].corr(method='spearman')
```

```python
sns.set_theme(style='white')
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=0.5, cbar_kws = {'label': 'Spearman Correlation'})
heatmap.set_title('Correlation Heatmap')
plt.show()
```

```python
books[books['description'].isna() |
      books['num_pages'].isna() |
      books['average_rating'].isna() |
      books['published_year'].isna()]
```

```python
books_no_missing = books[~books['description'].isna() &
                         ~books['num_pages'].isna() &
                         ~books['average_rating'].isna() &
                         ~books['published_year'].isna()].copy()
books_no_missing
```

```python
books_no_missing['categories'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
books_no_missing['words_in_description'] = books_no_missing['description'].str.split().str.len()
books_no_missing
```

```python
books_no_missing.loc[books_no_missing['words_in_description'].between(1, 4), 'description']
```

```python
books_no_missing.loc[books_no_missing['words_in_description'].between(25, 34), 'description']
```

```python
books_no_missing_25_words = books_no_missing[books_no_missing['words_in_description'] >= 25].copy()
books_no_missing_25_words
```

```python
books_no_missing_25_words['title_and_subtitle'] = (
    np.where(books_no_missing_25_words['subtitle'].isna(), books_no_missing_25_words['title'],
             books_no_missing_25_words[['title', 'subtitle']].astype(str).agg(': '.join, axis=1))
)
books_no_missing_25_words
```

```python
books_no_missing_25_words['tagged_description'] = books_no_missing_25_words[['isbn13', 'description']].astype(str).agg(' '.join, axis=1)
books_no_missing_25_words
```

```python
(books_no_missing_25_words
 .drop(['subtitle', 'missing_description', 'age_of_book', 'words_in_description'], axis=1)
 .to_csv('books_cleaned.csv', index=False)
 )
```
