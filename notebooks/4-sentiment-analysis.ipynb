```python
import numpy as np
from tqdm import tqdm
```

```python
emotion_labels = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}


def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}
    for prediction in predictions:
        for item in prediction:
            per_emotion_scores[item["label"]].append(item["score"])
    return {
        label: np.max(scores) if scores else 0.0
        for label, scores in per_emotion_scores.items()
    }
```

```python
for i in range(10):
    isbn.append(books.loc[i, "isbn13"])

    description = books.loc[i, "description"]
    sentences = [s.strip() for s in description.split(".") if s.strip()]
    if sentences:
        predictions = classifier(sentences, batch_size=16, truncation=True)
        max_scores = calculate_max_emotion_scores(predictions)
    else:
        max_scores = {label: 0.0 for label in emotion_labels}

    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
emotion_scores
```

```python
for i in tqdm(range(len(books))):
    isbn.append(books.loc[i, "isbn13"])

    description = books.loc[i, "description"]
    sentences = [s.strip() for s in description.split(".") if s.strip()]
    if sentences:
        predictions = classifier(sentences, batch_size=16)
        max_scores = calculate_max_emotion_scores(predictions)
    else:
        max_scores = {label: 0.0 for label in emotion_labels}

    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
for emotion in emotion_labels:
    print(emotion_scores[emotion][:10])
```

```python
emotion_df = pd.DataFrame(emotion_scores)
emotion_df["isbn13"] = isbn
emotion_df
```

```python
books = pd.merge(books, emotion_df, on="isbn13", how="left")
books
```

```python
books.to_csv('books_with_emotions.csv', index=False)
```
